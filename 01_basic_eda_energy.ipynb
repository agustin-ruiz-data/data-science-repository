```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset de ejemplo: producción de energía eólica
url = "https://raw.githubusercontent.com/owid/energy-data/master/owid-energy-data.csv"
df = pd.read_csv(url)

# Filtrar columnas de interés
df = df[['year', 'country', 'wind_share_energy']]

# Primer vistazo
print(df.head())
print(df.info())
print(df.describe())

# Gráfico simple: distribución de wind_share_energy
df['wind_share_energy'].dropna().hist(bins=20)
plt.title('Distribución % Energía Eólica')
plt.xlabel('%')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Promedio por país y top 10
df_country = df.dropna(subset=['wind_share_energy']).groupby('country')['wind_share_energy'].mean().sort_values(ascending=False).head(10)

# Gráfico de barras horizontales
plt.figure(figsize=(10,6))
sns.barplot(x=df_country.values, y=df_country.index)
plt.title('Top 10 países con mayor % de energía eólica')
plt.xlabel('% Energía Eólica')
plt.ylabel('País')
plt.show()
```
